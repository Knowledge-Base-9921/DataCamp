```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

# 1. Filter the data for movies released in the 1990s

```python
netflix_df.columns
```

```python
netflix_df.head(2)
```

```python
netflix_df_movie = netflix_df[netflix_df["type"]=="Movie"]
```

```python
netflix_df_movie_90s = netflix_df_movie[(netflix_df_movie["release_year"]>=1990) & (netflix_df_movie["release_year"]<2000)]
```

# 2. Find the most frequent movie duration

```python
netflix_df_movie_90s.head(2)
```

```python
plt.hist(netflix_df_movie_90s["duration"], bins=20)

plt.title("Frequent Movie Durations")
plt.xlabel("duration (minutes)")
plt.ylabel("count")

plt.show()
```

The most frequent movie duration in the 1990s is `100` minutes

# 3. Count the number of short action movies from the 1990s

```python
netflix_df.head(2)
```

```python
netflix_df_movie_90s_action = netflix_df_movie_90s[netflix_df_movie_90s["genre"]=="Action"]
```

```python
short_movie_count = 0
for lab, row in netflix_df_movie_90s_action.iterrows():
    if row["duration"] < 90:
        short_movie_count += 1

print(short_movie_count)
```
